```python
from gantt import *
```

```python
category_names = ['CPU1', 'CPU2', 'I/O Printer', 'I/O disk']

# results dict is scheduling data
# take 'P1' as example:
# since there are 4 categories: 'CPU1', 'CPU2', 'I/O Printer', 'I/O disk'
# so 'P1': [[(0, 80)], [(180, 40)], [(80, 100)], []] means:
# 'CPU1' from 0 to 0+80=80
# 'CPU2' from 180 to 180+40=220
# 'I/O Printer' from 80 to 80+100=180
# no 'I/O disk'
results = {
    'P1': [[(0, 80)], [(180, 40)], [(80, 100)], []],
    'P2': [[(80, 180), (330, 20)], [], [(260, 70)], []],
    'P3': [[], [(0, 130), (220, 50)], [], [(130, 50)]]
}

# program arrival time
arrival_t = [0, 10, 0]
```

```python
gantt(category_names, results, arrival_t).show()
```
